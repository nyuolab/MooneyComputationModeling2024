```python
fsaverage = datasets.fetch_surf_fsaverage("fsaverage7")
```

```python
import cortex
cortex.download_subject('fsaverage')
```

```python
left_data = surface.vol_to_surf("/Users/chris/projects/Nature2023MooneyScripts/results/encoding_significance/repetition-all/thresholded_pmap.nii", fsaverage["pial_left"])
right_data = surface.vol_to_surf("/Users/chris/projects/Nature2023MooneyScripts/results/encoding_significance/repetition-all/thresholded_pmap.nii", fsaverage["pial_right"])

vertex_data = cortex.Vertex(np.concatenate([left_data, right_data]), 'fsaverage')
fig = cortex.quickshow(vertex_data,
                        with_colorbar=True,
                        with_rois=False,
                        with_labels=False,
                        with_curvature=True,
                        curvature_contrast=0.5,
                        curvature_brightness=0.5,
                        curvature_threshold=True,
)
```
